```python
# =========================================
# YOLOv8 TRAINING FOR COTTON BALL GROWTH STAGE DETECTION
# =========================================

# Install YOLOv8
!pip install ultralytics -q

# =========================================
# IMPORTS
# =========================================

import os
from ultralytics import YOLO

# =========================================
# DATASET PATHS
# =========================================

DATASET_PATH = "refer_to_readme_for_dataset_description"

TRAIN_IMAGES = os.path.join(DATASET_PATH, "train/images")
VAL_IMAGES   = os.path.join(DATASET_PATH, "valid/images")
TEST_IMAGES  = os.path.join(DATASET_PATH, "test/images")

# =========================================
# CREATE data.yaml
# =========================================

data_yaml = f"""
path: {DATASET_PATH}

train: train/images
val: valid/images
test: test/images

names:
  0: cotton_ball
"""

with open("/kaggle/working/data.yaml", "w") as f:
    f.write(data_yaml)

print("✅ data.yaml created!")

# =========================================
# LOAD YOLOv8 MODEL
# =========================================

# Small/Fast Model
model = YOLO("yolov8n.pt")

# Better Accuracy Options:
# YOLO("yolov8s.pt")
# YOLO("yolov8m.pt")

# =========================================
# TRAIN MODEL
# =========================================

model.train(
    data="refer_readme",

    epochs=50,
    imgsz=640,

    batch=16,

    device="0,1",

    workers=2,

    optimizer="AdamW",

    lr0=1e-3,

    pretrained=True,

    patience=10,

    project="/kaggle/working",

    name="cotton_ball_growth_phase_yolov8",

    save=True,

    verbose=True
)

# =========================================
# VALIDATE MODEL
# =========================================

metrics = model.val()

print("\n========== VALIDATION METRICS ==========")

print(metrics)

# =========================================
# TEST INFERENCE
# =========================================

results = model.predict(
    source=TEST_IMAGES,
    conf=0.25,
    save=True
)

print("\n✅ Predictions completed!")

# =========================================
# EXPORT MODEL
# =========================================

model.export(format="onnx")

print("\n✅ ONNX model exported!")

# =========================================
# IMPORTANT OUTPUT FILES
# =========================================

print("\n========== IMPORTANT FILES ==========")

print("\nBest Weights:")
print("/kaggle/working/cotton_ball_yolov8/weights/best.pt")

print("\nLast Weights:")
print("/kaggle/working/cotton_ball_yolov8/weights/last.pt")

print("\nTraining Results:")
print("/kaggle/working/cotton_ball_yolov8/results.png")

print("\nConfusion Matrix:")
print("/kaggle/working/cotton_ball_yolov8/confusion_matrix.png")

print("\nPredictions:")
print("/kaggle/working/cotton_ball_yolov8/predict")

print("\nONNX Export:")
print("/kaggle/working/cotton_ball_yolov8/weights/best.onnx")
```
